```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from socket import timeout
from datetime import datetime 
import re
```

```python
url_ali = 'https://aliexpress.ru/category/202060516/tablets.html'
#url_ali = 'https://aliexpress.ru/category/202001502/mini-pc.html'
```

```python
df = pd.DataFrame()
```

```python
def if_not_none(item):
    #print(item)
    if item is not None:
        return item.text
    else:
        return None
```

```python
def parse_ali(link, back = ''):
    print(link)
    global df
    # link - ссылка на страницу
    # back - ссылка на возврат на страницу назад в случае ошибки парсера
    
    html_ali = urlopen(link, timeout=20)

    if back =='':
        back = link+"?page=" # шаблон возвратной ссылки
    #print(back)
    bs_ali = BeautifulSoup(html_ali.read(), 'html.parser')
    #print('bs_ali')
    
    page_active = bs_ali.find('a', {'class':'ali-kit_Base__base__104pa1 ali-kit_Base__default__104pa1 ali-kit_Link__link__cmgtoz ali-kit_Link__size-s__cmgtoz ali-kit_Link__default__cmgtoz SearchPagination_SearchPagination__paginationButton__wjhu3 SearchPagination_SearchPagination__paginationButtonActive__wjhu3'})
    if page_active != None:
        print('page_active', page_active.text)
    else:
        print('page_active - None')
    
    # Выбераем информацию о продуктах со страницы
    for item in bs_ali.find_all('a', {'class':'product-snippet_ProductSnippet__description__tusfnx'}):
        row_dict = {}
        item_soup = BeautifulSoup(str(item), 'html.parser')
        row_dict['title'] = [item_soup.find('div', {'class':'product-snippet_ProductSnippet__name__tusfnx'}).text]
        price = item_soup.find('div', {'class':'snow-price_SnowPrice__mainM__1ehyuw'}).text
        row_dict['price'] = re.sub(r' ', '', re.sub(r'руб.','',price))
        row_dict['score'] = [if_not_none(item_soup.find('div', {'class':'product-snippet_ProductSnippet__score__tusfnx'}))]
        sold = if_not_none(item_soup.find('div', {'class':'product-snippet_ProductSnippet__sold__tusfnx'}))
        if sold != None:
            row_dict['sold'] = [int(i) for i in str.split(sold) if i.isdigit()]
        else:
            row_dict['sold'] = None
        row_dict['store'] = [item_soup.find('div', {'class':'product-snippet_ProductSnippet__caption__tusfnx'}).text]

        df = pd.concat([df,pd.DataFrame(data=row_dict)], ignore_index=True)
        
    # ищем ссылку на следующую страницу
    next_page = None
    try:
        next_page = bs_ali.find(text = 'След. стр.').parent
        print(next_page.text, next_page.attrs['href'])
    except :
        print('Ссылка на следующую страницу не найдена. Похоже, дошли до конца.')
    # если нашли, то переходим дальше(с отработкой ошибки)
    if next_page != None:
        number_page = next_page.attrs['href'][next_page.attrs['href'].rfind('=')+1:]
        if next_page.attrs['href'][1:4] == 'who':
            print('oops', number_page, back+number_page)
            parse_ali(back+number_page, back)
        else:
            parse_ali('https://aliexpress.ru'+next_page.attrs['href'], back)
    html_ali.close()
```

```python
parse_ali(url_ali)
```
